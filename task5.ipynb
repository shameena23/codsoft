<h3>visualization of  dataset</h3>

```python
fraud = df[df['Class'] == 1]
nonfraud = df[df['Class'] == 0].sample(len(fraud))  # balance sample

balanced = pd.concat([fraud, nonfraud])

plt.figure(figsize=(18, 12))
for i, feature in enumerate(['V1', 'V2', 'V3', 'V4', 'V10', 'V11', 'V12', 'V14']):
    plt.subplot(3, 3, i+1)
    sns.kdeplot(balanced[balanced['Class'] == 0][feature], label='Non-Fraud', fill=True)
    sns.kdeplot(balanced[balanced['Class'] == 1][feature], label='Fraud', fill=True)
    plt.title(f'Distribution: {feature}')
    plt.legend()

plt.tight_layout()
plt.show()sns.countplot(x='Class', data=df, palette=['#1f77b4', '#ff7f0e'])
plt.title('Transaction Class Distribution')
plt.xticks([0, 1], ['Non-Fraud (0)', 'Fraud (1)'])
plt.ylabel('Count')
plt.show()
```

```python
counts = df['Class'].value_counts()
labels = ['Non-Fraud (0)', 'Fraud (1)']
colors = ['#1f77b4', '#ff7f0e']

plt.figure(figsize=(6,4))
plt.bar(labels, counts.values, color=colors)
for i, val in enumerate(counts.values):
    plt.text(i, val + 200, f'{val} ({val / df.shape[0] * 100:.4f}%)', ha='center')
plt.title('Fraud vs Non-Fraud Transactions')
plt.ylabel('Transaction Count')
plt.show()
```

<h4>Spliting data into train and test dataset</h4>

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=23)
```

```python
len(x_test)
```

<h3>Model training</h3>

<h4>Logistic Regression</h4

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test).astype(int)
y_pred
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.7, color='blue')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')  # ideal line
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Actual vs Predicted Sales')
plt.show()
```

```python
bins = np.arange(0, 11, 1)  
y_test_binned = pd.cut(y_test, bins=bins, labels=bins[:-1], include_lowest=True)
y_pred_binned = pd.cut(y_pred, bins=bins, labels=bins[:-1], include_lowest=True)
conf_mat = pd.crosstab(y_test_binned, y_pred_binned)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='YlGnBu', linewidths=0.5)
plt.xlabel('Predicted Rating Bin')
plt.ylabel('Actual Rating Bin')
plt.title('Binned Ratings Heatmap')
plt.show()
```

```python
model.score(x_test,y_test)
```

<h3>Model Accuracy</h3>

```python
Accuracy=model.score(x_test,y_test)
```

```python
print(f"The model achieved {Accuracy*100} % accuracy on the test set.")
```
